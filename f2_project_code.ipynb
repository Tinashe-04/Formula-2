```python
# Create a function that creates a new column called TIME IN SECONDS that converts the TIME column to seconds
def time_to_seconds(i):
    """
    This function converts the string from the TIME column to seconds

    Args: i (str): The string in the TIME column

    Returns: (float): The string in the TIME column converted to seconds

    Example: time_to_seconds('1:02:03.456') returns 3723.456000
             time_to_seconds('12:03.456') returns 723.456000   
    """
    # Strip the string of any whitespace
    # If the string has 11 characters and the first character is 1 or 2, then the hour is 1 or 2 and convert the hour to seconds
    # Otherwise if the string has 9 characters, then the hour is 0
    # Otherwise special case 1: '1:02.01.52'
    # Otherwise special case 2: '1:02:0.898'
    # Otherwise special case 3: '1:07:24.88'
    # Split the string from the decimal point of the seconds onwards to the rightmost number
    # Convert the minutes to seconds by type casting and multiplying by 60
    # Convert the seconds to seconds by type casting
    # Add the hour, minute and second to get the total time in seconds and type cast back to string
    # Add the string that was initially split to the rightmost number to get the microseconds
    # Return the total time in seconds as a float

    i = i.strip()

    try:
        # if len(i) == 11 and i.startswith(('1', '2')):                
        #     hour = int(i[0]) * 3600
        #     j = i[7:]
        #     minute = int(i[2:4]) * 60
        #     second = int(i[5:7])        
        
        # elif len(i) == 9:
        #     hour = 0
        #     j = i[5:]
        #     minute = int(i[0:2]) * 60
        #     second = int(i[3:5])

        # else:


        # new_prefix = hour + minute + second
        # new_time = str(new_prefix) + j
        if len(i) == 9:
            h = 0
            m, s, ms = map(int, i.split('.:'))

        elif len(i) == 10 and i[-3] == '.':
            i  += '0'
            h, m, s, ms = map(int, i.split(':.'))

        else:
            h, m, s, ms = map(int, i.split(':.'))

        new_time = (h * 3600) + (m * 60) + s + (ms / 1000)
        
    except ValueError:
        print(f"Invalid time format: {i}")
        new_time = i
    
    return new_time

df['TIME IN SECONDS'] = df['TIME'].apply(time_to_seconds)
```

```python
# use plotly for plotting cities that host races

# adjust the appearance of the second 'Sakhir' string by removing the space
df['CIRCUIT'] = df['CIRCUIT'].apply(lambda x: x.strip() if x == 'Sakhir ' else x)
df['CIRCUIT'].unique()
```

```python
### use plotly for plotting cities that host races

# map each city to a column for its country
# create a function that makes a dictionary to map each city as its key to a country value
def to_country(city):
    '''
    Purpose: to convert a city to a country
    Inputs: city, string that denotes the name of the city to be converted
    Returns: the name of the country to which that city belongs
    '''

    # create a list of countries for each city in the circuit list
    country_list = ['Bahrain', 'Spain', 'Monaco', 'Azerbaijan', 'Austria', 'United Kingdom', 'Hungary', 'Belgium', 'Italy',
    'Spain', 'United Arab Emirates', 'France', 'Russia', 'Italy', 'Saudi Arabia', 'Italy', 'Netherlands', 'Australia']

    # create a dictionary to map each city to the country name
    country = {df['CIRCUIT'].unique()[i]: country_list[i] for i in range(0, df['CIRCUIT'].nunique())}

    # return correct value of country name
    return country[city]

# apply the function to the data
df['COUNTRY'] = df['CIRCUIT'].apply(to_country)
df
```
